Load miRNA dari 3 Subtype Kidney

```python
import pandas as pd
import os

base_path = "/content/drive/MyDrive/TCGA-KICH"

# Load miRNA per subtype
mirna_kich = pd.read_csv(f"{base_path}/TCGA-KICH.mirna.tsv", sep="\t", index_col=0)
mirna_kirc = pd.read_csv(f"{base_path}/TCGA-KIRC.mirna.tsv", sep="\t", index_col=0)
mirna_kirp = pd.read_csv(f"{base_path}/TCGA-KIRP.mirna.tsv", sep="\t", index_col=0)

print("KICH:", mirna_kich.shape)
print("KIRC:", mirna_kirc.shape)
print("KIRP:", mirna_kirp.shape)
```

Harmonisasi miRNA (Intersection Feature)

```python
# Ambil intersection miRNA
common_mirna = (
    set(mirna_kich.index)
    & set(mirna_kirc.index)
    # set(mirna_kirc.index)
    & set(mirna_kirp.index)
)

print("Common miRNA:", len(common_mirna))

# Subset data
mirna_kich = mirna_kich.loc[sorted(common_mirna)]
mirna_kirc = mirna_kirc.loc[sorted(common_mirna)]
mirna_kirp = mirna_kirp.loc[sorted(common_mirna)]
```

Tambahkan Label Subtype (Ground Truth)

```python
mirna_kich_T = mirna_kich.T
mirna_kirc_T = mirna_kirc.T
mirna_kirp_T = mirna_kirp.T

mirna_kich_T["subtype"] = "KICH"
mirna_kirc_T["subtype"] = "KIRC"
mirna_kirp_T["subtype"] = "KIRP"
```

Gabungkan Semua Subtype (Final Dataset)

```python
mirna_all = pd.concat(
    [mirna_kich_T, mirna_kirc_T, mirna_kirp_T],
    # [mirna_kirc_T, mirna_kirp_T],
    axis=0
)

print("Final miRNA dataset:", mirna_all.shape)
mirna_all.head()
```

Cleaning & Validasi Sample ID

```python
# Validasi ID TCGA
mirna_all = mirna_all[mirna_all.index.str.startswith("TCGA-")]

# Cek missing value
print("Missing values:", mirna_all.isna().sum().sum())

# Optional: isi NaN
mirna_all = mirna_all.fillna(0)
```

Encode Label (Siap Multiclass)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
mirna_all["label"] = le.fit_transform(mirna_all["subtype"])

print(dict(zip(le.classes_, le.transform(le.classes_))))
```

Simpan Dataset Final

```python
mirna_all.to_csv(f"{base_path}/mirna_multiclass_ready.csv")

print("Saved: mirna_multiclass_ready.csv")
```

Load & Pisahkan Feature – Label

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/TCGA-KICH/mirna_multiclass_ready.csv", index_col=0)

X = df.drop(columns=["subtype", "label"])
y = df["label"]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

Normalisasi

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

FS -> LASSO

```python
from sklearn.linear_model import LogisticRegression

lasso = LogisticRegression(
    penalty="l1",
    solver="saga",
    multi_class="multinomial",
    max_iter=5000
)

lasso.fit(X_scaled, y)

selected = abs(lasso.coef_).sum(axis=0) > 0
X_sel = X_scaled[:, selected]

print("Selected miRNA:", X_sel.shape[1])
```

Train – Test Split (STRATIFIED!)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_sel, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

20 miRNA

A. REDUKSI FITUR → AMBIL TOP-K (mis. 20 miRNA)

```python
import numpy as np

# Hitung importance dari LASSO
importance = np.abs(lasso.coef_).sum(axis=0)

# Ambil hanya fitur yang lolos LASSO
importance_sel = importance[selected]

# Urutkan
top_k = 20   # bisa diganti 20 / 50
top_idx = np.argsort(importance_sel)[-top_k:]

X_top = X_sel[:, top_idx]

print("Final selected miRNA:", X_top.shape)
```

B. RESHAPE UNTUK DEEP LEARNING (LSTM)

```python
X_lstm = X_top.reshape(X_top.shape[0], X_top.shape[1], 1)

print("LSTM input shape:", X_lstm.shape)
```

C. STRATIFIED K-FOLD CROSS VALIDATION

```python
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.utils import to_categorical

y_cat = to_categorical(y, num_classes=3)

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

D. MODEL DEEP LEARNING (LSTM MULTIKELAS)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

def build_lstm_model(input_shape):
    model = Sequential([
        LSTM(64, input_shape=input_shape),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dense(3, activation="softmax")
    ])

    model.compile(
        optimizer=Adam(learning_rate=0.001),
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

E. TRAINING + CROSS VALIDATION

```python
from sklearn.metrics import accuracy_score, matthews_corrcoef
import numpy as np

acc_scores = []
mcc_scores = []

for fold, (train_idx, test_idx) in enumerate(skf.split(X_lstm, y)):
    print(f"\nFold {fold+1}")

    X_train, X_test = X_lstm[train_idx], X_lstm[test_idx]
    y_train, y_test = y_cat[train_idx], y_cat[test_idx]

    model = build_lstm_model((X_train.shape[1], 1))

    model.fit(
        X_train, y_train,
        epochs=40,
        batch_size=32,
        verbose=0
    )

    y_pred = np.argmax(model.predict(X_test), axis=1)
    y_true = np.argmax(y_test, axis=1)

    acc = accuracy_score(y_true, y_pred)
    mcc = matthews_corrcoef(y_true, y_pred)

    acc_scores.append(acc)
    mcc_scores.append(mcc)

    print(f"Accuracy: {acc:.4f} | MCC: {mcc:.4f}")
```

F. HASIL AKHIR (UNTUK BAB IV)

```python
print("\n=== CROSS VALIDATION RESULT ===")
print(f"Average Accuracy: {np.mean(acc_scores):.4f}")
print(f"Average MCC: {np.mean(mcc_scores):.4f}")
```

```python
import numpy as np

# Ambil koefisien LASSO
coef_sum = np.abs(lasso.coef_).sum(axis=0)

# Ambil indeks fitur non-zero
selected_idx = np.where(selected)[0]

# Ranking fitur
ranked_idx = selected_idx[np.argsort(coef_sum[selected_idx])[::-1]]

# Pilih TOP 35
top_k = 35
top_features_idx = ranked_idx[:top_k]

# Ambil nama miRNA
feature_names = X.columns
top_miRNAs = feature_names[top_features_idx]

print("Top 35 miRNA terpilih:")
print(top_miRNAs.tolist())
```

SIAPKAN DATA UNTUK DEEP LEARNING (35 FITUR)

```python
X_dl = X_scaled[:, top_features_idx]
```

MODEL DEEP LEARNING (MLP / LSTM-style)

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

y_cat = to_categorical(y, num_classes=3)

def build_model(input_dim):
    model = Sequential([
        Dense(64, activation="relu", input_dim=input_dim),
        Dropout(0.3),
        Dense(32, activation="relu"),
        Dropout(0.3),
        Dense(3, activation="softmax")
    ])
    model.compile(
        optimizer="adam",
        loss="categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model
```

CONFUSION MATRIX PER FOLD (Stratified K-Fold)

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold = 1
acc_scores = []

for train_idx, test_idx in skf.split(X_dl, y):
    print(f"\nFold {fold}")

    X_train, X_test = X_dl[train_idx], X_dl[test_idx]
    y_train, y_test = y_cat[train_idx], y[test_idx]

    model = build_model(X_dl.shape[1])
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=0)

    y_pred = np.argmax(model.predict(X_test), axis=1)

    acc = accuracy_score(y_test, y_pred)
    acc_scores.append(acc)

    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt="d",
                xticklabels=le.classes_,
                yticklabels=le.classes_,
                cmap="Blues")
    plt.title(f"Confusion Matrix - Fold {fold}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    fold += 1

print("Average Accuracy:", np.mean(acc_scores))
```

PLOT LEARNING CURVE (TRAIN vs VALIDATION)

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X_dl, y_cat,
    test_size=0.2,
    stratify=y,
    random_state=42
)

model = build_model(X_dl.shape[1])

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=100,
    batch_size=32,
    verbose=0
)

# Plot
plt.figure(figsize=(8,4))
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Learning Curve Deep Learning (35 miRNA)")
plt.legend()
plt.grid()
plt.show()
```

MCC (Matthews Correlation Coefficient) PER FOLD

```python
from sklearn.metrics import matthews_corrcoef

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

fold = 1
acc_scores = []
mcc_scores = []

for train_idx, test_idx in skf.split(X_dl, y):
    print(f"\nFold {fold}")

    X_train, X_test = X_dl[train_idx], X_dl[test_idx]
    y_train_cat = y_cat[train_idx]
    y_test = y[test_idx]

    model = build_model(X_dl.shape[1])
    model.fit(X_train, y_train_cat, epochs=50, batch_size=32, verbose=0)

    y_pred = np.argmax(model.predict(X_test), axis=1)

    acc = accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    acc_scores.append(acc)
    mcc_scores.append(mcc)

    print(f"Accuracy Fold {fold}: {acc:.4f}")
    print(f"MCC Fold {fold}: {mcc:.4f}")

    fold += 1

print("\n=== RATA-RATA ===")
print("Average Accuracy:", np.mean(acc_scores))
print("Average MCC:", np.mean(mcc_scores))
```

PERBANDINGAN 20 vs 35 vs 50 FITUR

```python
def evaluate_k_features(k):
    idx = ranked_idx[:k]
    X_k = X_scaled[:, idx]

    accs, mccs = [], []

    for train_idx, test_idx in skf.split(X_k, y):
        X_train, X_test = X_k[train_idx], X_k[test_idx]
        y_train_cat = y_cat[train_idx]
        y_test = y[test_idx]

        model = build_model(k)
        model.fit(X_train, y_train_cat, epochs=40, batch_size=32, verbose=0)

        y_pred = np.argmax(model.predict(X_test), axis=1)

        accs.append(accuracy_score(y_test, y_pred))
        mccs.append(matthews_corrcoef(y_test, y_pred))

    return np.mean(accs), np.mean(mccs)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import StratifiedKFold

# Define a helper function to evaluate for a given k and return lists of scores for each fold (using LSTM model)
def evaluate_lstm_k_features_detailed_local(k):
    idx = ranked_idx[:k]
    X_k_selected = X_scaled[:, idx]

    # Reshape X_k_selected for LSTM input
    X_lstm_k = X_k_selected.reshape(X_k_selected.shape[0], X_k_selected.shape[1], 1)

    accs, precs, recs, f1s = [], [], [], []

    # Re-initialize skf to ensure consistency, though it should be globally available
    skf_local = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    for train_idx, test_idx in skf_local.split(X_lstm_k, y): # y is Series, X_lstm_k is ndarray
        X_train_fold, X_test_fold = X_lstm_k[train_idx], X_lstm_k[test_idx]
        y_train_cat_fold = y_cat[train_idx]
        y_test_fold = y.iloc[test_idx] # Use .iloc for Series to prevent FutureWarning

        model = build_lstm_model((X_train_fold.shape[1], 1))
        model.fit(X_train_fold, y_train_cat_fold, epochs=40, batch_size=32, verbose=0)
        y_pred_fold = np.argmax(model.predict(X_test_fold), axis=1)

        accs.append(accuracy_score(y_test_fold, y_pred_fold))
        precs.append(precision_score(y_test_fold, y_pred_fold, average='macro', zero_division=0))
        recs.append(recall_score(y_test_fold, y_pred_fold, average='macro', zero_division=0))
        f1s.append(f1_score(y_test_fold, y_pred_fold, average='macro', zero_division=0))

    return accs, precs, recs, f1s

# Populate the individual lists for each k by calling the helper function
acc_list_20, prec_list_20, rec_list_20, f1_list_20 = evaluate_lstm_k_features_detailed_local(20)
acc_list_35, prec_list_35, rec_list_35, f1_list_35 = evaluate_lstm_k_features_detailed_local(35)
acc_list_50, prec_list_50, rec_list_50, f1_list_50 = evaluate_lstm_k_features_detailed_local(50)

# Construct the results dictionary using the newly created lists
results = {
    "20 miRNA": {
        "acc": acc_list_20,
        "prec": prec_list_20,
        "rec": rec_list_20,
        "f1": f1_list_20
    },
    "35 miRNA": {
        "acc": acc_list_35,
        "prec": prec_list_35,
        "rec": rec_list_35,
        "f1": f1_list_35
    },
    "50 miRNA": {
        "acc": acc_list_50,
        "prec": prec_list_50,
        "rec": rec_list_50,
        "f1": f1_list_50
    }
}
```
